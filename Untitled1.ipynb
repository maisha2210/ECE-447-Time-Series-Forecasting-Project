```python
import numpy as np                      # Numerical operations
import pandas as pd                     # Data handling

import matplotlib.pyplot as plt         # Plotting graphs

from sklearn.model_selection import TimeSeriesSplit   # Time-based cross-validation
from sklearn.linear_model import LinearRegression     # Linear Regression model
from sklearn.ensemble import RandomForestRegressor    # Random Forest model
from sklearn.preprocessing import StandardScaler      # Feature scaling

from xgboost import XGBRegressor        # XGBoost model
```

```python
#LINEAR REGRESSION

tscv = TimeSeriesSplit(n_splits=5)   # create 5 time-based splits

lr_mae, lr_rmse = [], []   # Lists to store errors

for train_index, test_index in tscv.split(X): # Split dataset into train and test
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    scaler = StandardScaler()  # Standardize features
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    
    model_lr = LinearRegression() # Train model
    model_lr.fit(X_train_scaled, y_train)
    
    y_pred_lr = model_lr.predict(X_test_scaled) # Predict
    
    lr_mae.append(np.mean(np.abs(y_test - y_pred_lr))) # Compute errors
    lr_rmse.append(np.sqrt(np.mean((y_test - y_pred_lr)**2)))

print("Linear Regression MAE:", np.mean(lr_mae))
print("Linear Regression RMSE:", np.mean(lr_rmse))
```

```python
#RANDOM FOREST

rf_mae, rf_rmse = [], []
for train_index, test_index in tscv.split(X):
    
    X_train, X_test = X.iloc[train_index], X.iloc[test_index] # Split data
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    model_rf = RandomForestRegressor(n_estimators=100, random_state=42) # Train model 
    model_rf.fit(X_train, y_train)
    
    y_pred_rf = model_rf.predict(X_test) # Predict
    
    rf_mae.append(np.mean(np.abs(y_test - y_pred_rf))) # Compute errors
    rf_rmse.append(np.sqrt(np.mean((y_test - y_pred_rf)**2)))

print("Random Forest MAE:", np.mean(rf_mae))
print("Random Forest RMSE:", np.mean(rf_rmse))
```

```python
#XGBOOST

xgb_mae, xgb_rmse = [], []

for train_index, test_index in tscv.split(X):
    
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]  # Split data
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    model_xgb = XGBRegressor(n_estimators=100, learning_rate=0.1) # Train model 
    model_xgb.fit(X_train, y_train)

    y_pred_xgb = model_xgb.predict(X_test) # Predict
    
    xgb_mae.append(np.mean(np.abs(y_test - y_pred_xgb))) # Compute errors
    xgb_rmse.append(np.sqrt(np.mean((y_test - y_pred_xgb)**2)))

print("XGBoost MAE:", np.mean(xgb_mae))
print("XGBoost RMSE:", np.mean(xgb_rmse))
```
